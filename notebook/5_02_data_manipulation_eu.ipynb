**1.2. Pivot Longer, egitura txukuntzeko**

```r
bpg_txukun1 <- bpg_urtea_zabal %>%
               pivot_longer(
               cols = `2020`:`2024`,
               names_to = "urtea",
               values_to = "bpg")
bpg_txukun1
```

```r
bpg_txukun2 <- bpg_lurraldea_zabal %>%
              pivot_longer(
              cols = EAE:Gipuzkoa,
              names_to = "lurraldea",
              values_to = "bpg")
bpg_txukun2
```

Beste adibide bat

```r
errenta_oso_zabal <- tibble(
  sexua = c("Guztira", "Gizonak", "Emakumeak"),

  EAE_2020 = c(22343, 27422, 17667),
  EAE_2021 = c(23313, 28360, 18659),
  EAE_2022 = c(25061, 30356, 20180),
  EAE_2023 = c(26271, 31641, 21322),

  Araba_2020 = c(21516, 26236, 17050),
  Araba_2021 = c(22373, 27148, 17853),
  Araba_2022 = c(23660, 28471, 19109),
  Araba_2023 = c(24921, 29886, 20238)
)

errenta_oso_zabal
```

```r
errenta_txukun2 <- errenta_oso_zabal %>%
  pivot_longer(
    cols = -sexua,
    names_to = c("lurraldea", "urtea"),
    names_sep = "_",
    values_to = "errenta"
  )

errenta_txukun2
```

Benetako kasu bat

 Giza garapenaren indizeko datuak erabiliko ditugu:
 https://hdr.undp.org/sites/default/files/2025_HDR/HDR25_Composite_indices_complete_time_series.csv

```r
url <- "https://raw.githubusercontent.com/uxue-sudupe/R-tailerrak/main/data/HDR_Composite_indices_complete_time_series.csv"

hdi <- read.csv(url)

hdi
```

```r
# Zutabe asko ditu, herrialdeen izenak eta giza garapenaren indizea dutenak bakarrik hartuko ditugu

hdi_bakarrik <- hdi %>%
  select(country, hdi_1990:hdi_2023)

head(hdi_bakarrik)
```

```r
hdi_txukun <-  hdi_bakarrik %>% pivot_longer(
    cols = starts_with("hdi_"),
    names_to = "urtea",
    names_prefix = "hdi_",
    values_to = "ggi"
  )

hdi_txukun
```

**1.2. PIVOT_WIDER()**

 Orain alderantziz egingo dugu

 Longer -> Wider
 Wider -> Longer



```r

hdi_berriz_zabal <- hdi_txukun %>%
  pivot_wider(
    names_from = urtea,
#   names_prefix = "hdi_",
    values_from = ggi
  )

hdi_berriz_zabal
```

**2. JOIN-AK**

Join bat egitea: Bi taula elkartu nahi ditut komunean duten gako baten bidez

```r
superheroiak <- tibble(
  id = c(1, 2, 3),
  izena = c("Batman",
            "Wonder Woman",
            "Superman"))
superheroiak
```

```r
botereak <- tibble(
  id = c(1, 2, 4),
  boterea = c(
    "Dirua",
    "Lazo magikoa",
    "Mailu erraldoia"))

botereak
```


**2.1 LEFT JOIN**

Gogoratzeko: "Ez galdu ezkerreko taulako inor"


```r

superheroiak %>%
  left_join(botereak, by = "id")
```


**2.2 INNER JOIN**

Hemen: Bi taulatan daudenak bakarrik azaldu

```r
superheroiak %>%
  inner_join(botereak, by = "id")
```

**2.3 RIGHT JOIN**

Oraingoan: Eskuineko taulako guztiak mantendu

```r

superheroiak %>%
  right_join(botereak, by = "id")
```

**2.4 FULL JOIN**

Denak sartzen dira

```r

superheroiak %>%
  full_join(botereak, by = "id")
```
